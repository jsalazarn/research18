```python
import pandas as pd
import numpy as np
from pandas import Series, DataFrame
import os
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
sns.set()
%matplotlib notebook
```

```python
os.chdir('D:\\Python Examples')
```

```python
koval = pd.read_excel('AllFields.xlsx', 'Koval')
```

```python
koval.head()
```

```python
koval.drop(['PDF'], axis=1, inplace=True)
```

```python
klc = koval.loc[:, ['(1-C)/C', '(1-F)/F']][koval['Field'] == 'Little Creek'].copy()
```

```python
kreb = koval.loc[:, ['(1-C)/C', '(1-F)/F']][koval['Field'] == 'REB'].copy()
```

```python
klc[klc < 0] = np.nan
kreb[kreb < 0] = np.nan
```

```python
fig, ax = plt.subplots(2, 1, sharex=True)
ax[0].plot(klc['(1-C)/C'], klc['(1-F)/F'], 'ro', alpha=0.2)
ax[1].plot(kreb['(1-C)/C'], kreb['(1-F)/F'], 'bs', alpha=0.2)

ax[0].set(xscale='log', yscale='log', xlabel='(1-C)/C', ylabel='(1-F)/F',
         xlim=(1e-4, 1e4), ylim=(1e-8, 1e2),
         title='Little Creek');

ax[1].set(xscale='log', yscale='log', ylabel='(1-F)/F',
         ylim=(1e-8, 1e2))

ax[0].xaxis.grid(True, which='minor', linestyle='-', linewidth=0.5,)
ax[1].xaxis.grid(True, which='minor', linestyle='-', linewidth=0.5,)

ax[0].yaxis.grid(True, which='minor', linestyle='-', linewidth=0.5,)
ax[1].yaxis.grid(True, which='minor', linestyle='-', linewidth=0.5,);
```

```python
fig, ax = plt.subplots()
ax.plot(klc['(1-C)/C'], klc['(1-F)/F'], 'ro', alpha=0.1, label='Little Creek')
ax.plot(kreb['(1-C)/C'], kreb['(1-F)/F'], 'bs', alpha=0.1, label='REB')

ax.set(xscale='log', yscale='log', xlabel='(1-C)/C', ylabel='(1-F)/F',
         xlim=(1e-4, 1e4), ylim=(1e-8, 1e2));
ax.set_title('Koval Plots', weight='bold', size=22)

ax.legend()

ax.xaxis.grid(True, which='minor', linestyle='-', linewidth=0.5,)
ax.xaxis.grid(True, which='minor', linestyle='-', linewidth=0.5,)
```
